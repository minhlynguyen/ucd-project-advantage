```python
from pgeocode import Nominatim
import pandas as pd

# Create a Nominatim object for the United States
nomi = Nominatim('us')

# Get the unique zip codes
unique_zip = df['Address ZIP'].unique()

# Create an empty DataFrame to store the results
location_df = pd.DataFrame(columns=['ZIP', 'Latitude', 'Longitude'])

for zip in unique_zip:
    # Get location data for each unique zip code
    location = nomi.query_postal_code(str(zip))
    
    # Create a DataFrame from the results
    new_row = pd.DataFrame({'ZIP': [zip], 'Latitude': [location.latitude], 'Longitude': [location.longitude]})
    
    # Concatenate the new row to the existing DataFrame
    location_df = pd.concat([location_df, new_row], ignore_index=True)

# Print the resulting DataFrame
print(location_df)
```

```python
location_df
```

```python
# Convert the ZIP code to int in both DataFrames to ensure they can match correctly
df['Address ZIP'] = df['Address ZIP'].astype(int)
location_df['ZIP'] = location_df['ZIP'].astype(str).apply(lambda x: x.split('.')[0]).astype(int)


# Merge df with location_df
merged_df = pd.merge(df, location_df, how='left', left_on='Address ZIP', right_on='ZIP')

merged_df
```

```python
# Use fillna to fill missing 'Latitude' and 'Longitude' values in df using merged_df
merged_df['Latitude_x'] = merged_df['Latitude_x'].fillna(merged_df['Latitude_y'])
merged_df['Longitude_x'] = merged_df['Longitude_x'].fillna(merged_df['Longitude_y'])

# Drop the extra columns
merged_df = merged_df.drop(columns=['Latitude_y', 'Longitude_y', 'ZIP'])

# Check the result
merged_df
```

```python
merged_df = merged_df.rename(columns={'Longitude_x': 'Longitude', 'Latitude_x': 'Latitude'})
merged_df
```

```python
# check the rows without Longitude and Latitude.

without_long_lat = merged_df[merged_df['Longitude'].isnull() | merged_df['Latitude'].isnull()]

without_long_lat
```

# do the point-in-polygon operation to match longitude and latitude points to specific taxi zones 

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
```

```python


df['Coordinates'] = df.apply(lambda row: Point(row.Longitude, row.Latitude), axis=1)

gdf_business = gpd.GeoDataFrame(df, geometry='Coordinates')  
```

```python
gdf_business
```

```python
# Read the geojson file
taxi_zone = gpd.read_file('NYC Taxi Zones.geojson')
taxi_zone
```

```python
gdf_taxi_zone = gpd.GeoDataFrame(taxi_zone, geometry='geometry')
gdf_taxi_zone
```

```python
# Ensure both GeoDataFrames are using the same CRS
gdf_business = gdf_business.set_crs("EPSG:4326")
gdf_taxi_zone = gdf_taxi_zone.set_crs("EPSG:4326")

# Perform spatial join with the 'predicate' parameter
result = gpd.sjoin(gdf_business, gdf_taxi_zone, how="inner", predicate='intersects')
```

```python
result
```
